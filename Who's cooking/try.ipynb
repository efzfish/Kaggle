```python
cuislist = []
for i in range(totlen):
    cuislist.append(data[i]['cuisine'])

cuis = list(set(cuislist))    
print len(cuis), len(cuislist)
```

```python
mat = {}
for i in cuis:
    for j in ingr:
        mat[(i,j)] = 0

for i in range(totlen):
    for j in range(len((data[i]['ingredients']))):
        mat[(data[i]['cuisine'],data[i]['ingredients'][j].encode('utf-8'))] += 1
```

```python
rarity = {}
for i in cuis:
    rarity[i] = 0

for i in range(totlen):
    tmprar = 0
    for j in data[i]['ingredients']:
        tmprar += ing[j.encode('utf-8')]
    rarity[(data[i]['cuisine'])] += tmprar
```

```python
# data[0] as an example
tmpd = {}
for i in cuis:
    tmpd[i] = 0
    for j in data[0]['ingredients']:
        if j in ingr:
            tmpd[i] += mat[(i,j)]
    
    tmpd[i] = tmpd[i] / rarity[i]

print tmpd

sorted_tmpd = [(k,v) for v,k in sorted([(v,k) for k,v in tmpd.items()])]
la = sorted_tmpd[-1][0]
print la
```

```python
# Now i'll get the results!

myresult = {}
for i in range(100):
    myresult[i] = 'NaN'

for num in range(100):
    tmpd = {}
    for i in cuis:
        tmpd[i] = 0
        for j in data[num]['ingredients']:
            if j in ingr:
                tmpd[i] += mat[(i,j)]

        tmpd[i] = tmpd[i] / rarity[i]
    
    sorted_tmpd = [(k,v) for v,k in sorted([(v,k) for k,v in tmpd.items()])]
    myresult[num] = sorted_tmpd[-1][0]
    
print myresult
```

```python
ct = 0
for i in range(100):
    if myresult[i] == data[i]['cuisine']:
        ct += 1
        
print ct
```

```python
jsonFile2 = open('test.json', 'r')
testdata = json.load(jsonFile2)

myresult = {}
for i in range(100):
    myresult[i] = 'NaN'

for num in range(100):
    tmpd = {}
    for i in cuis:
        tmpd[i] = 0
        for j in testdata[num]['ingredients']:
            if j in ingr:
                tmpd[i] += mat[(i,j)]

        tmpd[i] = tmpd[i] / rarity[i]
    
    sorted_tmpd = [(k,v) for v,k in sorted([(v,k) for k,v in tmpd.items()])]
    myresult[num] = sorted_tmpd[-1][0]
    
print myresult
```
